```python
with open("the-verdict.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
```

```python
print("total number of characters:", len(raw_text))
print(raw_text[:100]) 
```

```python
import re 

preprcoessed_text = re.split(r'([,.:;?_!()\'\"]|--|\s)', raw_text)
preprcoessed_text = [item.strip() for item in preprcoessed_text if item.strip()]
```

```python
preprcoessed_text[:10]
```

```python
print(len(preprcoessed_text))
```

```python
vocablulary = sorted(set(preprcoessed_text))
vocablulary_size = len(vocablulary)

print("vocabulary size:", vocablulary_size)
```

```python
vocablulary[:100]
```

```python
tokenid = {token:id for id, token in enumerate(vocablulary)}
```

```python
tokenid["<|unk|>"] = len(tokenid)  # Add unknown token to vocabulary
```

```python
for i, item in enumerate(tokenid.items()):
    print(item)
    if i >= 10:
        break
```

```python
class simpleTokenizerv2:
    def __init__(self, tokenid):
        self.strtoid = tokenid
        self.idtostr = {id:token for token, id in tokenid.items()}
    
    def encode(self, text):
        preprcoessed_text = re.split(r'([,.:;?_!()\'\"]|--|\s)', text)
        preprcoessed_text = [item.strip() for item in preprcoessed_text if item.strip()]
        preprcoessed_text = [item if item in self.strtoid 
                             else "<|unk|>" for item in preprcoessed_text]
        ids = [self.strtoid[token] for token in preprcoessed_text]

        return ids
    
    def decode(self, ids):
        text=" ".join([self.idtostr[id] for id in ids])
        text=re.sub(r'\s([,.:;?_!()\'\"])', r'\1', text)
        return text
```

```python
tokennizer = simpleTokenizerv2(tokenid)
encoded = tokennizer.encode("""helloooo world! this is a test.""")
tokennizer.decode(encoded)
```
